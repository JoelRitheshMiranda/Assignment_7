```python
# Read the following data set:
# https://archive.ics.uci.edu/ml/machine-learning-databases/adult/
# Rename the columns as per the description from this file:
# https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.names

# Task:
# Create a sql db from adult dataset and name it sqladb
# 1. Select 10 records from the adult sqladb
# 2. Show me the average hours per week of all men who are working in private sector
# 3. Show me the frequency table for education, occupation and relationship, separately
# 4. Are there any people who are married, working in private sector and having a masters degree
# 5. What is the average, minimum and maximum age group for people working in different sectors
# 6. Calculate age distribution by country
# 7. Compute a new column as 'Net-Capital-Gain' from the two columns 'capital-gain' and 'capital-loss'

import numpy as np
import pandas as pd
from pandas import DataFrame, Series
import sqlite3

ds = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data', header=None, sep=', ', engine='python')
ds.head()
```

```python
ds.columns = ['age', 'workclass', 'fnlwgt', 'education', 'education-num', 'marital_status', 'occupation', 'relationship', 'race', 'sex', 'capital_gain', 'capital_loss', 'hours_per_week', 'native_country', 'class']
ds.head()
```

```python
# Create a sql db from adult dataset and name it sqladb

connection = sqlite3.connect('sqladb')
cursor = connection.cursor()
ds.to_sql('adult', connection, if_exists='replace')
connection.commit()
```

```python
# 1. Select 10 records from the adult sqladb

data = pd.read_sql("SELECT * FROM adult limit 10", connection)
data
```

```python
# 2. Show me the average hours per week of all men who are working in private sector

#cursor.execute("SELECT avg(hours_per_week) Average FROM adult where trim(sex) = 'Male' and trim(workclass) = 'Private'")
#myresult = cursor.fetchall()
#for x in myresult:
#    print(x)
data = pd.read_sql("SELECT AVG(hours_per_week) Average FROM adult WHERE sex = 'Male' AND workclass = 'Private'", connection)
data
```

```python
# 3. Show me the frequency table for education, occupation and relationship, separately

print("=== Frequency table for Education ===") 
data_education = pd.read_sql("SELECT education, COUNT(education) frequency FROM adult GROUP BY education", connection)
data_education
```

```python
print("=== Frequency table for Occupation ===") 
data_occupation = pd.read_sql("SELECT occupation, COUNT(education) frequency FROM adult GROUP BY occupation", connection)
data_occupation
```

```python
print("=== Frequency table for Relationship ===") 
data_relationship = pd.read_sql("SELECT relationship, COUNT(education) frequency FROM adult GROUP BY relationship", connection)
data_relationship
```

```python
# 4. Are there any people who are married, working in private sector and having a masters degree

data_relationship = pd.read_sql("SELECT COUNT(marital_status) result FROM adult WHERE marital_status IN ('Married-civ-spouse', 'Married-spouse-absent' , 'Married-AF-spouse') AND workclass = 'Private' AND education = 'Masters'", connection)
print(data_relationship)
print("\nThere are", data_relationship["result"][0], "people who are married, working in private sector and having a masters degree")
```

```python
# 5. What is the average, minimum and maximum age group for people working in different sectors

data = pd.read_sql("SELECT workclass, AVG(age) average_age, MIN(age) minimum_age, MAX(age) maximum_age from adult group by workclass", connection)
data
```

```python
# 6. Calculate age distribution by country

data = pd.read_sql("SELECT native_country, AVG(age) average_age, MIN(age) minimum_age, MAX(age) maximum_age from adult group by native_country", connection)
data
```

```python
# 7. Compute a new column as 'Net-Capital-Gain' from the two columns 'capital-gain' and 'capital-loss'

data = pd.read_sql("SELECT capital_gain, capital_loss, (capital_gain - capital_loss) AS net_capital_gain FROM adult", connection)
data
```

```python
# Read the following data set:
# https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data

# Task:
# 1. Create an sqlalchemy engine using a sample from the data set
# 2. Write two basic update queries
# 3. Write two delete queries
# 4. Write two filter queries
# 5. Write two function queries

import sqlalchemy as db
import pandas as pd
from sqlalchemy import create_engine

dsSQLAlchemy = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data', header=None, sep=', ', engine='python')
dsSQLAlchemy
```

```python
columns = ['index', 'age', 'workclass', 'fnlwgt', 'education', 'education-num', 'marital_status', 'occupation', 'relationship', 'race', 'sex', 'capital_gain', 'capital_loss', 'hours_per_week', 'native_country', 'class']
dsSQLAlchemy.columns = ['age', 'workclass', 'fnlwgt', 'education', 'education-num', 'marital_status', 'occupation', 'relationship', 'race', 'sex', 'capital_gain', 'capital_loss', 'hours_per_week', 'native_country', 'class']
dsSQLAlchemy.head()
```

```python
# 1. Create an sqlalchemy engine using a sample from the data set

engine = db.create_engine('sqlite:///dbAdult.db')
connection = engine.connect()
dsSQLAlchemy.to_sql('Adult', engine, if_exists='replace')
metadata = db.MetaData()
tblAdult = db.Table('Adult', metadata, autoload=True, autoload_with=engine)
query = db.select([tblAdult])
resultset = connection.execute(query)
data = resultset.fetchall()
data
```

```python
# 2. Write two basic update queries

# Update Query
query = db.update(tblAdult).values(age = 50).where(tblAdult.columns.fnlwgt == 77516)
resultset = connection.execute(query)

# Display the updated data
query = db.select([tblAdult]).where(tblAdult.columns.fnlwgt == 77516)
resultset = connection.execute(query)
data = resultset.fetchall()
data
```

```python
# Update Query
query = db.update(tblAdult).values(capital_gain = 500).where(tblAdult.columns.sex == 'Male')
resultset = connection.execute(query)

# Display the updated data
query = db.select([tblAdult]).where(tblAdult.columns.sex == 'Male')
resultset = connection.execute(query)
data = resultset.fetchall()
data
```

```python
# 3. Write two delete queries

# Display the data
query = db.select([tblAdult]).where(tblAdult.columns.fnlwgt == 83311)
resultset = connection.execute(query)
data = resultset.fetchall()
print("=== Before Delete ===")
print(data)

# Delete Query
query = db.delete(tblAdult).where(tblAdult.columns.fnlwgt == 83311)
resultset = connection.execute(query)
print("\n\n=== Date Deleted ===")

# Display the data
query = db.select([tblAdult]).where(tblAdult.columns.fnlwgt == 83311)
resultset = connection.execute(query)
data = resultset.fetchall()
print("\n\n=== After Delete ===")
print(data)
```

```python
# 4. Write two filter queries

from sqlalchemy.orm import sessionmaker
Session = sessionmaker(bind=engine)
Session = sessionmaker()
Session.configure(bind=engine) 
session = Session()

data = session.query(tblAdult).filter(tblAdult.columns.age > 85).filter(tblAdult.columns.education == "Bachelors").all()
for row in data:
    print(row)
```

```python
from sqlalchemy import and_
data = session.query(tblAdult).filter(and_(tblAdult.columns.education == "Bachelors", tblAdult.columns.marital_status == "Married-civ-spouse")).all()
for row in data:
    print(row)
```

```python
# 5. Write two function queries

from sqlalchemy import func

data = session.query(tblAdult.columns.education,
                    func.avg(tblAdult.columns.age).label("average_age"),
                    func.min(tblAdult.columns.age).label("minimum_age"), 
                    func.max(tblAdult.columns.age).label("maximum_age")
                   ).group_by(tblAdult.columns.education).all()
for res in data:
    print(res)
```

```python
data = session.query(tblAdult.columns.occupation,
                    func.count(tblAdult.columns.occupation).label("count"),
                    func.sum(tblAdult.columns.capital_gain).label("total_capital_gain"),
                    func.sum(tblAdult.columns.capital_loss).label("total_capital_loss")
                   ).group_by(tblAdult.columns.occupation).all()
for res in data:
    print(res)
```
